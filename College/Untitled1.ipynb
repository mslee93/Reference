```python
cell_format = workbook.add_format() 
cell_format.set_border(1)
for i in range(len(data.index)):
    if data.loc[i,"Header"]!=current_header:
        
        if current_header!='':
        
            table = data.iloc[df_row:i-1,5:].as_matrix().tolist()
            # str(unichr(64))
            for row in table:
                start='A'+str(excel_row+1)
                worksheet.write_row(start, row, cell_format)
                excel_row += 1    
            # worksheet.add_table('A:F7', {'data': data})
    
        current_header = data.loc[i,"Header"]
        worksheet = workbook.add_worksheet(current_header[:30])
        excel_row = 2
        worksheet.write_string(excel_row, 0, current_header)
        excel_row += 2
        
        current_semiheader = data.loc[i,"SemiHeader"]
        worksheet.write_string(excel_row, 0, current_semiheader)
        excel_row += 2
        
        df_row = i
        
        
    elif data.loc[i,"SemiHeader"]!=current_semiheader:
        
        for row in table:
            start='A'+str(excel_row+2)
            worksheet.write_row(start, row, cell_format)
            excel_row += 1   
        
        excel_row +=2
        
        current_semiheader = data.loc[i,"SemiHeader"]
        worksheet.write_string(excel_row, 0, current_semiheader)
        excel_row += 2
        df_row = i
        
    else:
        pass

workbook.close()
```
