### Problem 1-3 Extracting data and filling it into a dictionary(Python data stucture)

#### Extract  countries and medals from "Athletics - Women" table and store it in dictionary.
*Please print parsed dictionary at the end of code*

#### Shape of Dictionary
__Include only once if athlete have received the same medal more than once. ex) 100m gold Bolt, 200m gold Bolt -> gold: Bolt__
- { *country_name*: { *medal type*: [*athlete name,* *athlete name,* *athlete name,* ...], ... }}
- e.g. {'Jamaica': {'Gold': ['Elaine Thompson', 'Shaunae Miller'...], 'Silver': [...], 'Bronze': [...] }, 'United States': ... }

*__hint: Use above problem's code to get Atheletics - Women table__*

```python
# extract headings and actual data and combine data into single dict

###YOUR CODE HERE###
ath_t=tables[2]
dic = {}
i=0
for row in ath_t.findAll('tr'):
    cells = row.findChildren('td')
    for cell in cells:
        content = cell.getText().strip().replace("\xa0", "").split("\n")
        if(content!=[''] and len(content)!=1 and content[1]!='details'):
            i+=1
            if(len(content)==2):
                if(content[1] not in dic.keys()):
                    dic[content[1]]={'Gold':[],'Silver':[],'Bronze':[]}
                if(i%3==1):
                    if(content[0] not in dic[content[1]]['Gold']):
                        dic[content[1]]['Gold'].append(content[0])
                elif(i%3==2):
                    if(content[0] not in dic[content[1]]['Silver']):
                        dic[content[1]]['Silver'].append(content[0])
                else:
                    if(content[0] not in dic[content[1]]['Bronze']):
                        dic[content[1]]['Bronze'].append(content[0])
            else:
                if(content[0]=='United States(USA)'):
                    content[0]='United States'
                elif(content[0]=='Jamaica(JAM)'):
                    content[0]='Jamaica'
                elif(content[0]=='Great Britain(GBR)'):
                    content[0]='Great Britain'
                if(content[0] not in dic.keys()):
                    dic[content[0]]={'Gold':[],'Silver':[],'Bronze':[]}
                if(i%3==1):
                    for j in range(1,len(content)):
                        if(content[j] not in dic[content[0]]['Gold']):
                            dic[content[0]]['Gold'].append(content[j])
                elif(i%3==2):
                    for j in range(1,len(content)):
                        if(content[j] not in dic[content[0]]['Silver']):
                            dic[content[0]]['Silver'].append(content[j])
                else:
                    for j in range(1,len(content)):
                        if(content[j] not in dic[content[0]]['Bronze']):
                            dic[content[0]]['Bronze'].append(content[j])
print(dic)
```

### Problem 3-1 Define a function to draw catter plot which contained trendline.
#### use .scatter from matplotlib library.
- please print *<b>line equation</b>* and *<b>MSE(mean squared error)</b>* for each plots
- matplotlib.pyplot.scatter: http://matplotlib.org/api/pyplot_api.html?highlight=scatter#matplotlib.pyplot.scatter
- you can find example of trendline on matplotlib very easily.

```python
from sklearn.metrics import mean_squared_error as m_s_e

def plot_2D_trendline(x_data, y_data, pred=None, x_label='X', y_label='Y', title=None):
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.scatter(x_data, y_data)
    
    if pred is not None:
        plt.plot(x_data, pred, color='red', linewidth=2)
    if title is not None:
        plt.title(title)
    
    # the line equation:
    #TODO
    z = np.polyfit(x_data, y_data, 1)
    p = np.poly1d(z)
    plt.plot(x_data,p(x_data),"r--")
    print ("y=%.6fx+(%.6f)"%(z[0],z[1]))
    
    #print Mean Squared Error
    y_pred=[]
    for x in x_data:
        y_pred.append(z[0]*x+z[1])
    mse = m_s_e(y_data,y_pred)
    print ("MeanSquaredError: ",mse)
    
    plt.show()

# Let's see if the function works.
plot_2D_trendline(np.random.randint(1000,size=100),np.random.randint(1000,size=100))    
```
