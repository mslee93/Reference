### 1) First Case

Columns include ['Gender', 'Married', 'Education', 'Self_Employed', 'Loan_Status']

#### - Gender

```python
data['Gender'].value_counts()
```

male and female로 구성 되어있어 Is_male, Is_female column을 만들어 0,1로 구성 한 후 기존 column 제거

```python
malelist = []
femalelist = []
for row in range(len(data.index)):
    if data['Gender'][row] == 'Male':
        malelist.append(1)
        femalelist.append(0)
    else:
        malelist.append(0)
        femalelist.append(1)
data['Is_male'] = malelist
data['Is_femlae'] = femalelist
data = data.drop('Gender', axis=1)
```

```python
data
```

#### - Married

```python
data['Married'].value_counts()
```

Yes or No 로 구성 되어있어 Yes 면 1 No 면 0으로 바꾼다

```python
data['Married'] = data['Married'].replace('Yes',1).replace('No',0)
```

```python
data
```

#### - Education

```python
data['Education'].value_counts()
```

Graduate or Not Graduate 로 구성 되어있어 Graduate 면 1 Not Graduate 면 0으로 바꾼다

```python
data['Education'] = data['Education'].replace('Graduate',1).replace('Not Graduate',0)
```

#### - Self_Employed

#### - Loan_Status

# 4. Run the Model & Feature Analyze 

## (1). Data split & Run Model

모델은 Random Forest를 쓰기로 정하고 이 모델로 부터 Feature의 중요도를 뽑아보려한다.<br>
우리에게 결과는 'Loan_Status'에 해당하는 column값이므로 이를 결과로 두고 나머지 값들을 Input으로 나누었다.

```python
data_x = data.loc[:, data.columns != 'Loan_Status']
data_y = data['Loan_Status']
```

```python
%matplotlib inline
from sklearn.ensemble import RandomForestClassifier
from sklearn_evaluation import plot
clf = RandomForestClassifier()
clf.fit(data_x,data_y)
```

## (2). Feature Importance

```python
plot.feature_importances(clf, top_n=5)
```

```python
values = sorted(zip(data_x.columns, clf.feature_importances_), key=lambda x: x[1] * -1)
for value in values:
    print(value)
```

## (3). Analyze the Feature Importance

가장영향을 많이 미치는 column은 다음 3개를 볼 수 있다.
- ('Credit_History', 0.24191630352712643)
- ('ApplicantIncome', 0.18689747824604003)
- ('LoanAmount', 0.18260467018861085)
<br>
이 중에서도 'Credit_History' 가 영향을 가장 많이 끼침을 볼 수 있다.<br>
따라서 우선 Credit_History column과 결과의 상관관계 및 상위 4개의 상관관계를 보려한다.

```python
data['Credit_History'].value_counts()
```

```python
resultdic={'Meet_Success':0, 'Meet_Fail':0, 'NotMeet_Success':0, 'NotMeet_Fail':0,}
for row in data.index:
    if data.loc[row,'Credit_History'] == 1:
        if data_y[row]==1:
            resultdic['Meet_Success']+=1
        else:
            resultdic['Meet_Fail']+=1
    else:
        if data_y[row]==1:
            resultdic['NotMeet_Success']+=1
        else:
            resultdic['NotMeet_Fail']+=1
print(resultdic)
```

```python
print('guideline에 맞을 경우 대출가능한 %: {:.2f}%'.format(resultdic['Meet_Success']/(resultdic['Meet_Success']+resultdic['Meet_Fail'])*100))
print('guideline에 맞지않을 경우 대출불가능한 %: {:.2f}%'.format(resultdic['NotMeet_Fail']/(resultdic['NotMeet_Success']+resultdic['NotMeet_Fail'])*100))
```

이를 보면 우선 guideline에 맞지 않은 경우 거의 대출이 불가함을 볼수 있고 맞는경우 80%의 확률로 대출이 가능함을 볼 수 있다.

다음으로 ApplicantIncome 과 LoanAmount를 보려한다.<br>
2가지 값 모두 수치이므로 각 column당 대출결과 상관관계를 보면 될듯 하다.

##### - ApplicantIncome

```python
data['ApplicantIncome'].describe()
```

ApplicantIncome를 전체 비율에서 -33%, 33%-66%, 66%- 구간으로 나누어 비교해보려 한다.<br>
그때의 값은 3000, 5500을 기준으로 두었다.

```python
appli_OK_resultdic = {'-3000':0, '3000-5500':0, '5500-':0}#4500':0, '4500-5500':0, '5500-6500':0, '6500-':0}
appli_NO_resultdic = {'-3000':0, '3000-5500':0, '5500-':0}
for row in data.index:
    if data.loc[row,'ApplicantIncome'] >= 5500:
        if data_y[row]==1:
            appli_OK_resultdic['5500-']+=1
        else:
            appli_NO_resultdic['5500-']+=1
    elif data.loc[row,'ApplicantIncome'] >= 3000:
        if data_y[row]==1:
            appli_OK_resultdic['3000-5500']+=1
        else:
            appli_NO_resultdic['3000-5500']+=1
    else:
        if data_y[row]==1:
            appli_OK_resultdic['-3000']+=1
        else:
            appli_NO_resultdic['-3000']+=1
print(appli_OK_resultdic)
print(appli_NO_resultdic)
```

두 값을 비교해본 결과 ApplicantIncome 값에 따라 Loan이 결정되는 것은 아닌것으로 보인다.<br>
또한 분포도 비슷한것을 보아서 대신 평균적으로 66%이상의 값이 대출을 받을 수 있는 것으로 보인다.

##### - LoanAmount

```python
data['LoanAmount'].value_counts()
```

이번에도 기준으로 100미만, 100이상-200미만, 200이상을 기준으로 잡았다.

```python
Loan_OK_resultdic = {'-100':0, '100-200':0, '200-':0}#4500':0, '4500-5500':0, '5500-6500':0, '6500-':0}
Loan_NO_resultdic = {'-100':0, '100-200':0, '200-':0}
for row in data.index:
    if int(data.loc[row,'LoanAmount']) >= 200:
        if data_y[row]==1:
            Loan_OK_resultdic['200-']+=1
        else:
            Loan_NO_resultdic['200-']+=1
    elif int(data.loc[row,'LoanAmount']) >= 100:
        if data_y[row]==1:
            Loan_OK_resultdic['100-200']+=1
        else:
            Loan_NO_resultdic['100-200']+=1
    else:
        if data_y[row]==1:
            Loan_OK_resultdic['-100']+=1
        else:
            Loan_NO_resultdic['-100']+=1
print(Loan_OK_resultdic)
print(Loan_NO_resultdic)
```

```python
print('100이하 대출가능 %: {:.2f}%'.format(Loan_OK_resultdic['-100']/(Loan_OK_resultdic['-100']+Loan_NO_resultdic['-100'])*100))
print('200이상 대출가능 %: {:.2f}%'.format(Loan_OK_resultdic['200-']/(Loan_OK_resultdic['200-']+Loan_NO_resultdic['200-'])*100))
```

두 값을 비교해본 결과 ApplicantIncome와 마찬가지로 값에 따라 강하게 Loan이 결정되는 것은 아닌것으로 보인다.
하지만 이번의 경우 영향이 크진않지만 확실히 대출이 많으면 적은사람에 비해 비율이 낮아짐을 볼 수 있다.
따라서 이로 볼때 대출값이 많은경우 적은 사람보다 대출가능성이 적어 진다.
